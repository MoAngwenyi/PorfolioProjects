```python
import pandas as pd
import numpy as np
import datetime
import os

print("Libraries imported successfully")
```

```python
df1 =pd.read_csv(r'C:\Users\User\OneDrive\Desktop\kaggle_set\Jan_2023.csv')
df2 =pd.read_csv(r'C:\Users\User\OneDrive\Desktop\kaggle_set\Feb_2023.csv')
df3 = pd.read_csv(r'C:\Users\User\OneDrive\Desktop\kaggle_set\March_2023.csv')
df4 = pd.read_csv(r'C:\Users\User\OneDrive\Desktop\kaggle_set\April_2023.csv')
print("CSV files successfully loaded")
```

```python
df1
```

```python
frames = (df1, df2, df3, df4)
main_df = pd.concat(frames)
main_df
```

```python
main_df.info()
```

```python
main_df.isnull().sum()
```

```python
main_df = main_df.dropna(how='any',axis=0)

print("Null values removed successfully.")

main_df.isnull().sum()
```

```python
main_df.duplicated().any()
```

```python
pd.options.mode.chained_assignment = None  # default='warn'

main_df["started_at"] = pd.to_datetime(main_df["started_at"])
main_df["ended_at"] = pd.to_datetime(main_df["ended_at"])

print("Typecast of variables changed successfully.") 
main_df.info()
```

```python
main_df["trip_duration"] = main_df["ended_at"] - main_df["started_at"]
main_df["trip_duration"] = pd.to_numeric(main_df["trip_duration"])/6e+10
```

```python
main_df["trip_duration"].describe(include = "all")
```

```python
# removing outliers by creating filters
outliers_trip_duration = (main_df["trip_duration"] < 1) | (main_df["trip_duration"] > 480)
outliers_trip_duration.value_counts()
```

```python
main_df.drop(main_df.index[outliers_trip_duration], inplace = True)

print("Inconsistent values deleted successfully.")
```

```python
main_df["trip_distance"] = np.sqrt(((main_df["end_lat"] - main_df["start_lat"])**2) + ((main_df["end_lng"] - main_df["start_lng"])**2))
main_df["trip_distance"] = main_df["trip_distance"] * 111.139
```

```python
main_df["start_hour"] = main_df["started_at"].dt.hour
main_df["start_hour"].unique()
```

```python
main_df["start_day_num"] = main_df["started_at"].dt.dayofweek
day_dict = {0:"Sunday", 1:"Monday", 2:"Tuesday", 3:"Wednesday", 4:"Thursday", 5:"Friday", 6:"Saturday"}
main_df["start_day_name"]= main_df["start_day_num"].apply(lambda y:day_dict[y])
```

```python
main_df["weekend"] = np.where(main_df["start_day_num"] > 5, True, False)
```

```python
main_df['start_month_num'] = main_df['started_at'].dt.month
main_df["start_month_num"].unique()
```

```python
month_dict = {1:"January", 2:"February", 3:"March", 4:"April"}
main_df["start_month_name"]= main_df["start_month_num"].apply(lambda y:month_dict[y])
main_df["start_month_name"].unique()
```

```python
main_df['start_year'] = main_df['started_at'].dt.year
main_df["start_year"].unique()
```

```python
main_df.to_csv('cyclistic_dataset.csv', index=False)

print("Dataset exported as CSV file successfully.")
```
